```python
#leer el archivo de top differentially expressed genes que se obtuvo de la base de datos asignarle nombre de GDE
GDE = pd.read_csv('Topdifferentiallyexpressedgenes.csv')
```

```python
#Para obtener el nombre de los 10 genes con el adj.P.Val m치s peque침o y los 10 genes con el adj.P.Val m치s grande de la columna Gene.title del archivo GDE
import pandas as pd
GDE_sorted_asc = GDE.sort_values('adj.P.Val', ascending=True)
GDE_sorted_desc = GDE.sort_values('adj.P.Val', ascending=False)
genes_with_smallest_adj_pval = GDE_sorted_asc['Gene.title'].head(10)
genes_with_largest_adj_pval = GDE_sorted_desc['Gene.title'].head(10)

```

```python
#Imprimir los resultados de los 10 genes con el adj.P.Val m치s peque침o 
print("Genes con el adj.P.Val m치s peque침o:")
print(genes_with_smallest_adj_pval)
```

```python
#Imprimir los resultados de los 10 genes con el adj.P.Val m치s grande
print("Genes con el adj.P.Val m치s grande:")
print(genes_with_largest_adj_pval)
```

```python
#Para generar un archivo llamado "top_GDE" que contenga los resultados tanto para los 10 genes con el adj.P.Value m치s peque침o como para los 10 genes con el adj.P.Value m치s grande, agregando en una columna sus valores de adj.P.Value, se puede usar el siguiente c칩digo:
# Ordenar el DataFrame por la columna "adj.P.Val" en orden ascendente y descendente
GDE_sorted_asc = GDE.sort_values('adj.P.Val', ascending=True)
GDE_sorted_desc = GDE.sort_values('adj.P.Val', ascending=False)

# Obtener los 10 genes con el adj.P.Val m치s peque침o y m치s grande junto con sus valores de adj.P.Val
top_genes = pd.concat([GDE_sorted_asc[['Gene.title', 'adj.P.Val']].head(10), GDE_sorted_desc[['Gene.title', 'adj.P.Val']].head(10)])

# Guardar los resultados en un archivo llamado "1top_GDE"
top_genes.to_csv('top_GDE.csv', index=False)

# Imprimir los resultados de los 10 genes con el adj.P.Val m치s peque침o y m치s grande
print("Genes con el adj.P.Val m치s peque침o y m치s grande:")
print(top_genes)
```

```python
#Hacer un gr치fico de volcan del archivo que se llama infiltrating ductal carcinoma vs normal, padjmenor0.05forVolcanoplot usando para el eje de las x la columna que se llama log2(fold change) y como eje de las y la columna que se llama -log10(Pvalue)
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
filename = "infiltratingductalcarcinomavsnormalpadjmenor0.05forVolcanoplot.csv"
data = pd.read_csv(filename)
#convertir los valores no n칰meros en valores NAN
data["adj.P.Val"] = pd.to_numeric(data["adj.P.Val"], errors='coerce')
# Generar el gr치fico de volcan
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data["log2(fold change)"], -1 * data[".-log10(Pvalue)"], color="grey", alpha=0.5)

# Seleccionar los puntos significativos
sig_genes = data[data["adj.P.Val"] < 0.05]
up_genes = sig_genes[sig_genes["log2(fold change)"] > 0]
down_genes = sig_genes[sig_genes["log2(fold change)"] < 0]

# Resaltar los puntos significativos
ax.scatter(up_genes["log2(fold change)"], -1 * up_genes[".-log10(Pvalue)"], color="red", alpha=0.8, label="Upregulated genes")
ax.scatter(down_genes["log2(fold change)"], -1 * down_genes[".-log10(Pvalue)"], color="blue", alpha=0.8, label="Downregulated genes")

# Configurar el gr치fico
ax.set_xlabel("log2(fold change)")
ax.set_ylabel(".-log10(Pvalue)")
ax.set_title("Volcano plot - Infiltrating ductal carcinoma vs normal")
ax.legend()

# Guardar el gr치fico
fig.savefig("graficadeVolcan_GED.png")

```

```python
#Diagrama de Venn
import pandas as pd
from matplotlib_venn import venn3

# Cargar el archivo en un dataframe
df = pd.read_csv("infiltratingductalcarcinomvsnormal_limma_padvmenor0.05forVennDiagram.csv")
#convertir los valores no n칰meros en valores NAN
df["adj.P.Val"] = pd.to_numeric(df["adj.P.Val"], errors='coerce')
df.dropna(subset=["adj.P.Val"], inplace=True)

# Crear conjuntos para cada columna
genes = set(df["Gene.symbol"])
upregulated = set(df[df["log2_fold_change"] > 0]["Gene.symbol"])
downregulated = set(df[df["log2_fold_change"] < 0]["Gene.symbol"])
significant = set(df[df["adj.P.Val"] < 0.05]["Gene.symbol"])

# Crear el diagrama de Venn
venn3([genes, upregulated, downregulated], ("Todos los genes", "Genes up-regulated", "Genes down-regulated"))
```

```python
#Gr치fica de barras din치mica con las funciones de GO para los genes sobreexpresados, se va a cargar el archivo llamado "Genes.menor.adj.p.val.GO.csv" se va usar en el eje de las y, las columnas llamadas "adj.P.Val" ,"GO.Function","GO.Process","GO.Component", en el eje de las x la columna llamada "Gene.symbol"
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo en un dataframe
df = pd.read_csv("Genes.menor.adj.p.val.GO.csv")

# Seleccionar columnas de inter칠s
df = df[["Gene.symbol", "adj.P.Val", "GO.Function", "GO.Process", "GO.Component"]]

# Ordenar genes por valor de adj.P.Val
df.sort_values("adj.P.Val", ascending=False, inplace=True)

# Configurar gr치fica
plt.figure(figsize=(10, 7))
plt.title("Funciones GO para genes sobreexpresados")
plt.xlabel("Genes")
plt.ylabel("adj.P.Val")
plt.xticks(rotation=90)

# Generar gr치fica de barras apiladas
bottom = 0
for col in ["GO.Function", "GO.Process", "GO.Component"]:
    df[col].fillna("", inplace=True)
    plt.bar(df["Gene.symbol"], df["adj.P.Val"], bottom=bottom, label=col)
    bottom += df["adj.P.Val"]

plt.legend()

# Guardar gr치fica
plt.savefig("grafica_funciones_genessobreexpresados_GO.png")
```
